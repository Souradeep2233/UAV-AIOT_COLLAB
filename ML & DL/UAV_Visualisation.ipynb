```python
data = {'Clear':110, 'Partially Cloudy':129, 'Rainy , Overcast':142,
        'Rainy , Partially Cloudy':14}
courses = list(data.keys())
values = list(data.values())
  
fig = plt.figure(figsize = (10, 5))
 
# creating the bar plot
plt.bar(courses, values, color ='orange',
        width = 0.4)
 
plt.xlabel("Conditions")
plt.ylabel("Days Frequency")
plt.title("Daily different Weather Conditions frequency")
plt.show()
```
